```python
import tensorflow as tf
```

```python
input_shape = (56,56,3)
enet = tf.keras.applications.efficientnet.EfficientNetB7(
    include_top=False,
    weights='imagenet',
    input_tensor=None,
    input_shape=input_shape,
    pooling=None,
    classes=1,
    classifier_activation='softmax')
```
